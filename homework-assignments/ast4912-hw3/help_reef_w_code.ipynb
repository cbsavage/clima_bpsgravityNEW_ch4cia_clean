```python
# import necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
nwt = np.loadtxt( 'fort.80' )
```

```python
num  = nwt[ :, 0 ]
conv = nwt[ :, 1 ]
```

```python
# plot Fortran data

plt.figure( figsize=(6,5) )
plt.plot( num, conv )

plt.xlabel( 'X', fontsize=10 )
plt.ylabel( 'Y', fontsize=10 )

# plt.ylim( [-6.0,-4.0] )
# plt.ylim( [-30,-3.0] )

plt.title( "Convergence by Newton's Method", fontsize=12 ) # - Fort 80 Edition", fontsize=12 )

plt.show()
```

```python
import math as meth
```

```python
reef3 = np.genfromtxt( 'fort.80', max_rows=5 )

# guess2 = reef2[ :, 0 ]
answ3  = reef3[ :, 1 ]

def function( x ):
    k = ( 1/4 ) * meth.sin( x ) + 5
    return k

x = np.linspace( -10, 10, 100 )
bee = function( x )

plt.figure( figsize=(6,5) )
plt.plot( answ3, bee )

plt.xlabel( 'Root', fontsize=10 )
plt.ylabel( 'Function', fontsize=10 )

# plt.ylim( [-6.0,-4.0] )
# plt.ylim( [-30,-3.0] )

plt.title( "Newton's Method", fontsize=12 ) # - Fort 80 Edition", fontsize=12 )

plt.show()
```
